```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

wine=pd.read_csv("winequality-red.csv",sep=";")
wine.info()
```

```python

X=wine.drop("quality",axis=1).values
y=wine["quality"].values
```

```python
X_FAcid=X[:,0].reshape(-1,1)
print(y.shape,X_FAcid.shape)
```

```python
plt.scatter(X_FAcid,y)
plt.xlabel("Fixed Acidity")
plt.ylabel("Quality")
plt.show()
```

```python
reg=LinearRegression()
reg.fit(X_FAcid,y)
pred=reg.predict(X_FAcid)
```

```python
plt.scatter(X_FAcid,y)
plt.plot(X_FAcid,pred,color="red")
plt.xlabel("Fixed Acidity")
plt.ylabel("Quality")
plt.show()
```

```python
score = reg.score(X_FAcid, y)
print("R-squared score:", score)
```
