```python
import pandas as pd
import matplotlib.pyplot as plt
# Load the datasets
train = pd.read_csv('Train.csv')
test = pd.read_csv('Test.csv')
display(train.head())
display(test.head())
```

```python
# Display the shape of each DataFrame
print("Shape of Train DataFrame:",train.shape)
print("Shape of Test DataFrame:",test.shape)
```

```python
# Summary statistics of both DataFrames
print("Summary Statistics for Train DataFrame:")
print(train.describe())
print("\nSummary Statistics for Test DataFrame:")
print(test.describe())
```

```python
# Check for null values in both DataFrames
print("Null Values in Train DataFrame:")
print(train.isnull().sum())
print("\nNull Values in Test DataFrame:")
print(test.isnull().sum())
```

```python
# Check for unique values in the species column
unique_species_train = train['Class'].unique()
print("\nUnique Species in Train DataFrame:",unique_species_train)

unique_species_test = test['Class'].unique()
print("\nUnique Species in Test DataFrame:",unique_species_test)
```

```python
# Separate the features and target variable in X and y
X=train.drop('Class',axis=1)  # Features
y=train['Class']                # Target variable
display(y)
display(X)
```

```python
from sklearn.tree import DecisionTreeClassifier
# Initialize the Decision Tree model
dt_model=DecisionTreeClassifier(random_state=42)

# Fit the model on the training data
dt_model.fit(X, y)
```

```python
# Perform predictions on the training data
train_pred = dt_model.predict(X)

# Check Train accuracy
from sklearn.metrics import accuracy_score

train_accuracy = accuracy_score(y,train_pred)
print("Train Accuracy:",train_accuracy)
```

```python
# Separate the test data into dependent and independent variables
X_test = test.drop('Class', axis=1)  # Features
y_test = test['Class']                # Target variable (if available)
display(y)
display(X)
```

```python
# Perform predictions on the test data
test_pred = dt_model.predict(X_test)
```

```python
# Check Test accuracy
test_accuracy = accuracy_score(y_test, test_pred)
print("Test Accuracy:", test_accuracy)
```

```python
# Importing train_test_split to create validation set
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(X, y, random_state = 10, stratify=y, test_size=0.25)
```

```python
train_accuracy = []
validation_accuracy = []
for depth in range(1,10):
    dt_model = DecisionTreeClassifier(max_depth=depth, random_state=10)
    dt_model.fit(X_train, y_train)
    train_accuracy.append(dt_model.score(X_train, y_train))
    validation_accuracy.append(dt_model.score(X_valid, y_valid))
```

```python
frame = pd.DataFrame({'max_depth':range(1,10), 'train_acc':train_accuracy, 'valid_acc':validation_accuracy})
frame
```

```python

plt.figure(figsize=(12,6))
plt.plot(frame['max_depth'], frame['train_acc'], marker='o',label='train_Acc')
plt.plot(frame['max_depth'], frame['valid_acc'], marker='o',label='Valid_Acc')
plt.xlabel('Depth of tree')
plt.ylabel('performance')
plt.legend()
```
